# download pdf files automatically

```python
import signal
import time
 
def set_timeout(num, callback):
    def wrap(func):
        def handle(signum, frame): 
            raise RuntimeError
 
        def to_do(*args, **kwargs):
            try:
                signal.signal(signal.SIGALRM, handle) 
                signal.alarm(num) 
                print('start alarm signal.')
                r = func(*args, **kwargs)
                print('close alarm signal.')
                signal.alarm(0)  
                return r
            except RuntimeError as e:
                callback()
 
        return to_do
 
    return wrap
 
def after_timeout():  
    print("Time out!")

# example
# @set_timeout(2, after_timeout) 
# def connect(): 
#     time.sleep(3)  
#     print('Finished without timeout.')
```

```python
import requests

@set_timeout(60*5, after_timeout)
def download_pdf(save_path,pdf_name,pdf_url):
    try:
        print("downloading %s.pdf" % (pdf_name))
        r = requests.get(pdf_url, verify=False)
        with open (save_path+pdf_name+".pdf", "wb") as code:
            code.write(r.content)
            print('%s.pdf download sucecssfully' % (pdf_name))
    except Exception as e:
        print("error downloading pdf from " + pdf_url)    
        print(e)
```

```python
import os
with open("./s&p500.text", "r") as f_obj:
    for pdf_url in f_obj:
        name = pdf_url.split("/")[-1].split(".pdf")[0].replace("\n", "")
        if not os.path.exists("./pdf_files/"+name+".pdf"):
            download_pdf("./pdf_files/", name, pdf_url)
```

# delete invalid pdf files

```python
import traceback
from PyPDF2 import PdfFileReader

def isValidPDF_pathfile(pathfile):
    bValid = True
    try:
        # PdfFileReader(open(pathfile, 'rb'))
        reader = PdfFileReader(pathfile)
        if reader.getNumPages() < 1:
            bValid = False
    except:
        bValid = False
        print(pathfile + " is invalid and will be removed!")
        # print("*" + traceback.format_exc())

    return bValid
```

```python
import os
import pandas as pd

input_path = "./pdf_files"
pdf_files = os.listdir(input_path)
for pdf_file in pdf_files:
    path = os.path.join(input_path, pdf_file)
    if not isValidPDF_pathfile(path):
        os.remove(path)
```
